```python
print("List operations called append, remove or pop can modify a list.")
print("Append adds an item to the end of a list")
print("Remove removes a specific element from the list")
print("Pop removes the last item of a list")

# Example
"""Operations for adding and removing elements from a list."""

# Create a list of students working on a project
student_list = ["Jamie", "Vicky", "DeShawn", "Tae"]
print(student_list)

# Another student joins the project. The student must be added to the list.
student_list.append("Ming")
print(student_list)

# "Jamie" withdraws from the project. Jamie must be removed from the list.
student_list.remove("Jamie")
print(student_list)

# Suppose "Ming" had to be removed from the list.
# A pop() operation can be used since Ming is last in the list.
student_list.pop()
print(student_list)

#Exercise
my_list = [2, 3, 5, 7, 9]
for i in range(0, len(my_list), 2):
    print(my_list[i], end= ' ')

"""Exercise Problem
Create a list of sports played on a college campus. 
The sports to be included are baseball, football, tennis, and table tennis.
Next, add volleyball to the list.
Next, remove "football" from the list and add "soccer" to the list.
Show the list contents after each modification."""

sports_list = ["baseball", "football", "tennis", "table tennis"]
print(sports_list)
sports_list.append("volleyball")
print(sports_list)
sports_list.remove("football")
print(sports_list)
sports_list.append("soccer")
print(sports_list)

# To print "found!" if soccer is found in the list:
for element in sports_list:
    if element == "soccer":
        print("found!")
        break
        
```

```python
print("When a list is an element inside another list, it is called a nested list")
print("Intuitively, it is clear that multidimensional data is an application of nested lists")

#example
matA = [[7,3,1],[4,9,2],[5,6,8]]
print(matA[2][1])
print(matA[0])

print("A nested loop can be used to iterate a list of lists")
"""Book example - Iterating a list-of-lists."""

# Create a list of numbers
list1 = [[1, 2, 3],[1, 4, 9],[1, 8, 27]]
# Iterating the list-of-lists
for row in list1:
  for num in row:
    print(num, end=" ")
  print()

my_list = [[7, 4, 5, 12],
          [24, 3, 9, 16],
          [12, 8, 91, -5]]
for row in my_list:
    for elem in row:
        print(elem)

for elem in my_list:
    print(elem)

for column_index in range(0, len(my_list[0])):
  for row_index in range (0, len(my_list)):
    print(my_list[row_index][column_index])
for row_index in range(0, len(my_list)):
  for column_index in range (0, len(my_list)):
    print(my_list[row_index][column_index])
  print()
for row_index in range(0, len(my_list)):
  for column_index in range (0, len(my_list[0])):
    print(my_list[row_index][column_index])

# Exercise to multiply two given matrices
# Create two lists for the matrices W and Z
matW = [[13,4,5],
    [2,-9,7],
    [7,3,19]]

matZ = [[2,1,5],
    [3,7,9],
    [-1,13,19]]

# Create a result matrix with zeros initially
result = [[0,0,0],[0,0,0],[0,0,0]]

# Calculate each element of the result matrix 
# The i and j for loops are used to go element by element
# The k for loop is used to calculate the product to be stored at each position in the result.
for i in range(len(matW)):
    for j in range(len(matZ[0])):  
       for k in range(len(matW)):
           result[i][j] += matW[i][k] * matZ[k][j]

# Print the result
print(result)
```
